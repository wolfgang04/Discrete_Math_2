### Task 1: Q1 Expected Value Calculation

The expected value \( E[y] \) is given by:

$$ E[y] = \sum y \cdot P(y) $$

Given the joint probability distribution formula:

$$ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $$

We need to show that $( E[y] = n \cdot p_i $).

$$ E[y] = \sum y \cdot P(y) = \sum\_{y_1,y_2,...,y_n} y_1 \cdot P(y_1) \cdot y_2 \cdot P(y_2) \cdot ... \cdot y_n \cdot P(y_n) $$

Since \( P(y) \) follows the given joint probability distribution formula, we can substitute it in:

$$ E[y] = \sum\_{y_1,y_2,...,y_n} y_1 \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} \cdot y_2 \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} \cdot ... \cdot y_n \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $$

Simplifying and using the fact that $( y_1 + y_2 + ... + y_n = n $):

$$ E[y] = \sum\_{y_1,y_2,...,y_n} n \cdot P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $$

Since the trials are independent and $( P_1 + P_2 + ... + P_n = 1 $), we can simplify further:

$$ E[y] = n \cdot \left( P_1 + P_2 + ... + P_n \right) = n \cdot 1 = n \cdot p_i $$

This shows that $( E[y] = n \cdot p_i $), which is the expected value.

$$ E[y] = n \cdot p_i $$

### Task 2: Probability Distribution and Expectation

#### Question 1: Probability Distribution

The probability distribution of the experiment can be derived from the given steps:

1. The experiment consists of $n$ identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$, and the probability of failure is equal to $q = (1 − p)$.
4. The trials are independent.

The random variable of interest is $Y$, the number of successes observed during the $n$ trials.

The probability distribution can be represented as:

$$ P(Y=y) = \binom{n}{y} p^y q^{n-y} $$

Where:
- $n$ is the number of trials,
- $y$ is the number of successes,
- $p$ is the probability of success on a single trial, and
- $q$ is the probability of failure on a single trial.

#### Question 2: Expectation Calculation

To show that the expectation of this probability distribution is $E[y] = \sum y \cdot P(y) = np$:

$$ E[y] = \sum_{y=0}^{n} y \cdot P(y) = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^y q^{n-y} $$

Using the binomial theorem, $p^y q^{n-y} = (p+q)^n$, we can simplify:

$$ E[y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} (p+q)^n $$

Since $(p+q)^n = 1^n = 1$, we have:

$$ E[y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} = np $$

This shows that $E[y] = np$, which is the expectation.

#### Question 3: Probability of Recovery

Given that 30% of all persons recover from the illness, and 10 people received the medication with 9 recovering, we want to find the probability of at least 9 out of 10 recovering if the medication is worthless.

Since the medication being worthless implies a 30% recovery rate by chance, we can use the binomial distribution formula to calculate the probability:

$$ P(\text{at least 9 recover}) = \sum_{y=9}^{10} \binom{10}{y} (0.3)^y (0.7)^{10-y} $$

#### Question 4: Probability of Defective Fuses

If a lot of 5000 electrical fuses contains 5% defectives and a sample of 5 fuses is tested, we want to find the probability of observing at least one defective fuse.

Using the complement rule, the probability of observing at least one defective fuse is equal to 1 minus the probability of observing no defective fuses:

$$ P(\text{at least one defective}) = 1 - P(\text{no defective}) = 1 - \binom{5}{0} (0.05)^0 (0.95)^5 $$

#### Question 1: Expected Value Formula

The expected value, or mean, of a probability distribution is calculated as:

$$ E[Y] = \sum_{y=0}^{\infty} y \cdot p(y) $$

For the given probability distribution $( p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $), the expected value can be expressed as:

$$ E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $$

This sum can be simplified using the fact that $( e^{-\lambda} $) and $( \frac{\lambda^y}{y!} $) are components of the Poisson distribution, where $( \lambda $) represents the mean. Therefore, the expected value of this distribution is $( E[Y] = \lambda $).

### Question 2: Probability Calculation

Given that the mean density of seedlings is five per square yard, the probability of a region not containing any seedlings can be calculated using the Poisson distribution formula:

$$ P(Y = 0) = \frac{\lambda^0}{0!} e^{-\lambda} = e^{-\lambda} $$

where $( \lambda = 5 $) in this case.

Substitute $( \lambda = 5 $) into the formula to find the probability:

$$ P(Y = 0) = e^{-5} \approx 0.00674 $$

Therefore, the probability that none of the regions will contain seedlings is approximately 0.00674.
