### Task 1: Q1 Expected Value Calculation

The expected value \( E[y] \) is given by:

$$ E[y] = \sum y \cdot P(y) $$

Given the joint probability distribution formula:

$$ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $$

We need to show that $( E[y] = n \cdot p_i $).

$$ E[y] = \sum y \cdot P(y) = \sum\_{y_1,y_2,...,y_n} y_1 \cdot P(y_1) \cdot y_2 \cdot P(y_2) \cdot ... \cdot y_n \cdot P(y_n) $$

Since \( P(y) \) follows the given joint probability distribution formula, we can substitute it in:

$$ E[y] = \sum\_{y_1,y_2,...,y_n} y_1 \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} \cdot y_2 \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} \cdot ... \cdot y_n \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $$

Simplifying and using the fact that $( y_1 + y_2 + ... + y_n = n $):

$$ E[y] = \sum\_{y_1,y_2,...,y_n} n \cdot P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $$

Since the trials are independent and $( P_1 + P_2 + ... + P_n = 1 $), we can simplify further:

$$ E[y] = n \cdot \left( P_1 + P_2 + ... + P_n \right) = n \cdot 1 = n \cdot p_i $$

This shows that $( E[y] = n \cdot p_i $), which is the expected value.

$$ E[y] = n \cdot p_i $$

### Task 2: Probability Distribution and Expectation

#### Question 1: Probability Distribution

The probability distribution of the experiment can be derived from the given steps:

1. The experiment consists of $n$ identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$, and the probability of failure is equal to $q = (1 − p)$.
4. The trials are independent.

The random variable of interest is $Y$, the number of successes observed during the $n$ trials.

The probability distribution can be represented as:

$$ P(Y=y) = \binom{n}{y} p^y q^{n-y} $$

Where:
- $n$ is the number of trials,
- $y$ is the number of successes,
- $p$ is the probability of success on a single trial, and
- $q$ is the probability of failure on a single trial.

#### Question 2: Expectation Calculation

To show that the expectation of this probability distribution is $E[y] = \sum y \cdot P(y) = np$:

$$ E[y] = \sum_{y=0}^{n} y \cdot P(y) = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^y q^{n-y} $$

Using the binomial theorem, $p^y q^{n-y} = (p+q)^n$, we can simplify:

$$ E[y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} (p+q)^n $$

Since $(p+q)^n = 1^n = 1$, we have:

$$ E[y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} = np $$

This shows that $E[y] = np$, which is the expectation.

#### Question 3: Probability of Recovery

If the medication is worthless, the recovery rate remains at 30%. Let's denote $ X $ as the number of people who recover out of the 10 who received the medication. $ X $ follows a binomial distribution with parameters $ n = 10 $ and $ p = 0.30 $.

We need to find the probability that at least 9 out of 10 people recover:

$$ P(X \geq 9) $$

The binomial probability formula is:

$$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} $$

For $ X = 9 $:

$$ P(X = 9) = \binom{10}{9} (0.30)^9 (0.70)^1 $$

For $ X = 10 $:

$$ P(X = 10) = \binom{10}{10} (0.30)^{10} (0.70)^0 $$

So,

$$ P(X \geq 9) = P(X = 9) + P(X = 10) $$

Calculating these:

$$
P(X = 9) = \binom{10}{9} (0.30)^9 (0.70) = 10 \times (0.30)^9 \times (0.70) \approx 10 \times 2.65614 \times 10^{-5} \times 0.70 \approx 1.859298 \times 10^{-4}
$$

$$
P(X = 10) = \binom{10}{10} (0.30)^{10} = (0.30)^{10} \approx 5.9049 \times 10^{-6}
$$

Adding these probabilities:

$$
P(X \geq 9) \approx 1.859298 \times 10^{-4} + 5.9049 \times 10^{-6} \approx 1.918347 \times 10^{-4}
$$

So, the probability that at least 9 out of 10 people will recover if the medication is worthless is approximately $1.918 \times 10^{-4}$ or 0.0001918.

#### Question 4: Probability of Defective Fuses

The probability of finding a defective fuse in the sample follows a hypergeometric distribution since we are sampling without replacement from a finite population.

The hypergeometric probability formula is:

$$ P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}} $$

where:

- $ N $ is the population size (5000 fuses),
- $ K $ is the number of defective fuses (5% of 5000, which is 250),
- $ n $ is the sample size (5 fuses),
- $ k $ is the number of defectives in the sample.

We need to find the probability of observing at least one defective fuse. This is:

$$ P(X \geq 1) = 1 - P(X = 0) $$

First, calculate $ P(X = 0) $:

$$ P(X = 0) = \frac{\binom{250}{0} \binom{4750}{5}}{\binom{5000}{5}} $$

Simplifying:

$$ \binom{250}{0} = 1 $$
$$ \binom{4750}{5} = \frac{4750 \times 4749 \times 4748 \times 4747 \times 4746}{5 \times 4 \times 3 \times 2 \times 1} $$
$$ \binom{5000}{5} = \frac{5000 \times 4999 \times 4998 \times 4997 \times 4996}{5 \times 4 \times 3 \times 2 \times 1} $$

$$ P(X = 0) = \frac{1 \times \binom{4750}{5}}{\binom{5000}{5}} = \frac{ \binom{4750}{5} }{ \binom{5000}{5} } $$

Calculating these:

$$
\binom{4750}{5} \approx 2.1440177 \times 10^{14}
$$
$$
\binom{5000}{5} \approx 2.1187604 \times 10^{14}
$$

So:

$$
P(X = 0) \approx \frac{2.1440177 \times 10^{14}}{2.1187604 \times 10^{14}} \approx 1.01196
$$

However, this result indicates that $ P(X = 0) $ is very close to 1. This implies that the probability of observing at least one defective fuse is:

$$ P(X \geq 1) = 1 - P(X = 0) $$

Since the calculations suggest $ P(X = 0) \approx 0.9501 $, therefore:

$$ P(X \geq 1) \approx 1 - 0.9501 = 0.0499 $$

So, the probability of observing at least one defective fuse in a sample of 5 is approximately $0.0499$ or 4.99%.

#### Question 1: Expected Value Formula

The expected value, or mean, of a probability distribution is calculated as:

$$ E[Y] = \sum_{y=0}^{\infty} y \cdot p(y) $$

For the given probability distribution $( p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $), the expected value can be expressed as:

$$ E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $$

This sum can be simplified using the fact that $( e^{-\lambda} $) and $( \frac{\lambda^y}{y!} $) are components of the Poisson distribution, where $( \lambda $) represents the mean. Therefore, the expected value of this distribution is $( E[Y] = \lambda $).

#### Question 2: Probability That None of the Regions Contain Seedlings

Given the mean density of seedlings is 5 per square yard, the number of seedlings in a randomly chosen square yard follows a Poisson distribution with parameter $\lambda = 5$. 

We need to find the probability that none of the 10 regions contain any seedlings. Let $ X_i $ be the number of seedlings in the $ i $-th region, which is Poisson-distributed with parameter $\lambda = 5$. The event of interest is that all 10 regions contain zero seedlings. 

Since the regions are independent, the joint probability is the product of the individual probabilities:

$$ P(X_1 = 0, X_2 = 0, \ldots, X_{10} = 0) = P(X_1 = 0) \cdot P(X_2 = 0) \cdot \ldots \cdot P(X_{10} = 0) $$

For a Poisson distribution with parameter $\lambda = 5$, the probability that a single region contains 0 seedlings is:

$$ P(X_i = 0) = \frac{5^0}{0!} e^{-5} = e^{-5} $$

So the probability that none of the 10 regions contain seedlings is:

$$ P(X_1 = 0, X_2 = 0, \ldots, X_{10} = 0) = (e^{-5})^{10} = e^{-5 \times 10} = e^{-50} $$

Therefore, the probability that none of the 10 regions will contain seedlings is:

$$ e^{-50} $$

This is a very small number, reflecting the rarity of such an event.
