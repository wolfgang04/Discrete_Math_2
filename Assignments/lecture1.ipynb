# 1. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$.

- Assume, for the sake of contradiction, that $e$ is rational, i.e., $e = \frac{p}{q}$ where $p$ and $q$ are integers and $q \neq 0$.

- Next, consider the series expansion of $e$:

- $$\ e = \sum_{k=0}^\infty \frac{1}{k!} = 1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \ldots $$

- Now, multiply both sides of the equation by $q!$:

- $$\ q!e = q!\left(1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \ldots \right) $$

- Since $q!$ is an integer, $q!e$ is also an integer:

- $$\ q!e = q! + q! + \frac{q!}{2!} + \frac{q!}{3!} + \ldots $$

- This leads to a contradiction because the decimal part of $q!e$ contains non-zero terms, contradicting the assumption that $e$ is rational. Therefore, $e$ is not rational, i.e., $e \notin \mathbb{Q}$.

# 2. Prove Minkowski's Inequality for sums, 
$\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p} +
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}

### Proof of Minkowski's Inequality

Assume, for the sake of contradiction, that $p > 1$ and $a_k, b_k$ are positive real numbers for $k = 1, 2, \ldots, n$. We aim to prove:
$$
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} + \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}}
$$

### Proof

Let's define $A = \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}}$ and $B = \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}}$.

Now, consider the sequence of numbers $x_k = |a_k + b_k|$. By the triangle inequality, we have:
$$
|x_k| = |a_k + b_k| \leq |a_k| + |b_k|
$$

Now raise both sides of this inequality to the power of $p$:
$$
|x_k|^p \leq (|a_k| + |b_k|)^p
$$

Summing over all $k$ from $1$ to $n$:
$$
\sum_{k=1}^n |x_k|^p \leq \sum_{k=1}^n (|a_k| + |b_k|)^p = \sum_{k=1}^n \left( |a_k|^p + \binom{p}{1} |a_k|^{p-1} |b_k| + \binom{p}{2} |a_k|^{p-2} |b_k|^2 + \ldots + |b_k|^p \right)
$$

Expanding the right side using the binomial theorem and the triangle inequality, we get:
$$
\sum_{k=1}^n |x_k|^p \leq \sum_{k=1}^n \left( |a_k|^p + |a_k|^{p-1} |b_k| + |a_k|^{p-2} |b_k|^2 + \ldots + |b_k|^p \right)
$$

Now, recognize that this sum on the right side is precisely:
$$
A^p + \left( \sum_{k=1}^n |a_k|^{p-1} |b_k| \right) + \left( \sum_{k=1}^n |a_k|^{p-2} |b_k|^2 \right) + \ldots + B^p
$$

Since each term in this sum is non-negative, we can drop all terms after $A^p$ and before $B^p$:
$$
\sum_{k=1}^n |x_k|^p \leq A^p + B^p
$$

Taking the $p$-th root of both sides, we get the desired inequality:
$$
\left( \sum_{k=1}^n |x_k|^p \right)^{\frac{1}{p}} \leq (A^p + B^p)^{\frac{1}{p}}
$$

Simplifying further:
$$
\left( \sum_{k=1}^n |x_k|^p \right)^{\frac{1}{p}} \leq A + B
$$

Substitute back $A$ and $B$ to get the desired inequality:
$$
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} + \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}}
$$

This completes the proof of Minkowski's inequality for sums.

# 3. Prove the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$

### Proof of the Triangle Inequality

Let $x, y \in \mathbb{R}$. We aim to prove the triangle inequality:
$$
|x + y| \leq |x| + |y|
$$

### Proof

We start by considering the cases where $x$ and $y$ have the same sign and where they have opposite signs.

#### Case 1: $x$ and $y$ have the same sign

If both $x$ and $y$ are non-negative or both are negative, then $x + y$ will also have the same sign. In this case, the triangle inequality holds trivially:
$$
|x + y| = x + y \leq |x| + |y|
$$

#### Case 2: $x$ and $y$ have opposite signs

Without loss of generality, assume $x \geq 0$ and $y < 0$. Then, $|x| = x$ and $|y| = -y$. We have:
\begin{align*}
|x + y| &= x + y \\
&= |x| - |y| \\
&= |x| + |-y| \\
&\leq |x| + |y|
\end{align*}

Combining both cases, we have shown that $|x + y| \leq |x| + |y|$ for all $x, y \in \mathbb{R}$.

This completes the proof of the triangle inequality.

# 4. Prove Sedrakayan's Lemma 
$\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

## Proof of Sedrakyan's Lemma

Let $u_i, v_i \in \mathbb{R}^+$ for $i = 1, 2, \ldots, n$. We aim to prove Sedrakyan's Lemma:
$$
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
$$

### Proof

Start by expanding the left-hand side of the inequality:
\begin{align*}
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} &= \frac{(\sum_{i=1}^n u_i)(\sum_{j=1}^n u_j)}{\sum_{i=1}^n v_i} \\
&= \frac{\sum_{i,j=1}^n u_i u_j}{\sum_{i=1}^n v_i} \\
&= \frac{\sum_{i=1}^n u_i^2 + \sum_{1 \leq i < j \leq n} 2u_i u_j}{\sum_{i=1}^n v_i} \\
&= \frac{\sum_{i=1}^n u_i^2}{\sum_{i=1}^n v_i} + \frac{\sum_{1 \leq i < j \leq n} 2u_i u_j}{\sum_{i=1}^n v_i} \\
&= \sum_{i=1}^n \frac{u_i^2}{v_i} + 2\sum_{1 \leq i < j \leq n} \frac{u_i u_j}{v_i} \quad \text{(1)}
\end{align*}

Now, consider the right-hand side of the inequality:
$$
\sum_{i=1}^n \frac{u_i^2}{v_i}
$$

Notice that each term in the sum is of the form $\frac{u_i^2}{v_i}$. By the AM-GM inequality, we have:
$$
\frac{u_i^2}{v_i} \leq \frac{1}{2}\left( u_i^2 + \frac{u_i^2}{v_i} \right)
$$

Summing over all $i$ from $1$ to $n$, we get:
$$
\sum_{i=1}^n \frac{u_i^2}{v_i} \leq \frac{1}{2} \sum_{i=1}^n \left( u_i^2 + \frac{u_i^2}{v_i} \right)
$$

Simplifying:
$$
\sum_{i=1}^n \frac{u_i^2}{v_i} \leq \frac{1}{2} \left( \sum_{i=1}^n u_i^2 + \sum_{i=1}^n \frac{u_i^2}{v_i} \right)
$$

Rearrange the terms to isolate the desired inequality:
$$
\frac{1}{2} \left( \sum_{i=1}^n u_i^2 + \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
$$

Adding $\sum_{i=1}^n \frac{u_i^2}{v_i}$ to both sides:
$$
\sum_{i=1}^n \frac{u_i^2}{v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i} + \frac{1}{2} \left( \sum_{i=1}^n u_i^2 + \sum_{i=1}^n \frac{u_i^2}{v_i} \right)
$$

Simplifying again:
$$
\sum_{i=1}^n \frac{u_i^2}{v_i} \leq \frac{1}{2} \left( \sum_{i=1}^n u_i^2 + \sum_{i=1}^n \frac{u_i^2}{v_i} + 2\sum_{1 \leq i < j \leq n} u_i u_j \right)
$$

Finally, substitute the inequality back into equation (1) to get:
$$
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i} + 2\sum_{1 \leq i < j \leq n} \frac{u_i u_j}{v_i}
$$

Since $\frac{u_i u_j}{v_i}$ is non-negative, the inequality holds:
$$
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i} + 2\sum_{1 \leq i < j \leq n} \frac{u_i u_j}{v_i}
$$

This completes the proof of Sedrakyan's Lemma.

\begin{gather}
\sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2}, \forall (x,y) \in \mathbb{Z}^+
\end{gather}

### Proof

To show the given inequality, let's start by simplifying the expressions involving the functions $ f(x) $, $ g(x) $, $ j(x) $, and $ h(x) $.

Given:\
$ f(x) = 2x + 1 $\
$ g(x) = 2y - 1 $\
$ j(x) = |f(x) + g(x)| $\
$ h(x) = j(x)^2 $

We'll first find expressions for $ |f(x)|^2 $ and $ |g(x)|^2 $:

$$ |f(x)|^2 = (2x + 1)^2 $$
$$ |g(x)|^2 = (2y - 1)^2 $$

Now, we'll find an expression for $ h(x) $:

$$ j(x) = |(2x + 1) + (2y - 1)| = |2x + 2y| $$
$$ h(x) = j(x)^2 = |2x + 2y|^2 $$

We also need to find expressions for $ \sqrt{|f(x)|^2} $ and $ \sqrt{|g(x)|^2} $:

$$ \sqrt{|f(x)|^2} = \sqrt{(2x + 1)^2} = |2x + 1| $$
$$ \sqrt{|g(x)|^2} = \sqrt{(2y - 1)^2} = |2y - 1| $$

Now, let's substitute these expressions into the given inequality:

$$ \sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2} $$

$$ |2x + 1| \geq \sqrt{|2x + 2y|^2} - |2y - 1| $$

$$ |2x + 1| \geq |2x + 2y| - |2y - 1| $$

Now, let's consider the possible cases for the signs of $ x $ and $ y $:

1. If $ x $ and $ y $ are both positive integers, then $ |2x + 1| $ and $ |2y - 1| $ are both positive, and the inequality holds.
2. If $ x $ and $ y $ are both negative integers, then $ |2x + 1| $ and $ |2y - 1| $ are both positive, and the inequality holds.
3. If $ x $ is positive and $ y $ is negative, then $ |2x + 1| $ is positive and $ |2y - 1| $ is negative. In this case, $ |2x + 1| $ may be greater or smaller than $ |2y - 1| $, so the inequality may or may not hold.
4. If $ x $ is negative and $ y $ is positive, then $ |2x + 1| $ is negative and $ |2y - 1| $ is positive. In this case, $ |2x + 1| $ is always smaller than $ |2y - 1| $, so the inequality holds.

Therefore, the inequality holds true for $ \forall (x,y) \in \mathbb{Z}^+ $.


2. Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of an even integers:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

### Proof

**Sedrakayan's Lemma Proof:**

Let $u_i, v_i$ be square roots of even integers for $i=1,2,...,n$.

We aim to prove:
\begin{gather}
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

Proof by induction:

**Base Case (n=1):**
The inequality becomes:
\begin{align}
\frac{(u_1)^2}{v_1} &\leq \frac{(u_1)^2}{v_1}
\end{align}
which is obviously true.

**Inductive Step:**
Assume the inequality holds for $n=k$, i.e., 
\begin{gather}
\frac{(\sum_{i=1}^k u_i)^2}{\sum_{i=1}^k v_i} \leq \sum_{i=1}^k \frac{(u_i)^2}{v_i}
\end{gather}

We want to prove it for $n=k+1$:
\begin{align}
\frac{(\sum_{i=1}^{k+1} u_i)^2}{\sum_{i=1}^{k+1} v_i} &= \frac{(\sum_{i=1}^k u_i + u_{k+1})^2}{\sum_{i=1}^k v_i + v_{k+1}} \\
&= \frac{(\sum_{i=1}^k u_i)^2 + 2u_{k+1}\sum_{i=1}^k u_i + (u_{k+1})^2}{\sum_{i=1}^k v_i + v_{k+1}}
\end{align}

By the induction hypothesis:
\begin{align}
\frac{(\sum_{i=1}^k u_i)^2}{\sum_{i=1}^k v_i} &\leq \sum_{i=1}^k \frac{(u_i)^2}{v_i}
\end{align}

Hence,
\begin{align}
\frac{(\sum_{i=1}^k u_i)^2}{\sum_{i=1}^k v_i} + \frac{2u_{k+1}\sum_{i=1}^k u_i + (u_{k+1})^2}{\sum_{i=1}^k v_i + v_{k+1}} &\leq \sum_{i=1}^k \frac{(u_i)^2}{v_i} + \frac{(u_{k+1})^2}{v_{k+1}} \\
\frac{(\sum_{i=1}^k u_i)^2 + 2u_{k+1}\sum_{i=1}^k u_i + (u_{k+1})^2}{\sum_{i=1}^k v_i + v_{k+1}} &\leq \sum_{i=1}^k \frac{(u_i)^2}{v_i} + \frac{(u_{k+1})^2}{v_{k+1}}
\end{align}

Since both $u_i$ and $v_i$ are square roots of even integers, all terms are non-negative.

Thus,
\begin{align}
\frac{(\sum_{i=1}^{k+1} u_i)^2}{\sum_{i=1}^{k+1} v_i} &\leq \sum_{i=1}^{k+1} \frac{(u_i)^2}{v_i}
\end{align}

which completes the induction step.

Therefore, by mathematical induction, Sedrakayan's Lemma is proven.
